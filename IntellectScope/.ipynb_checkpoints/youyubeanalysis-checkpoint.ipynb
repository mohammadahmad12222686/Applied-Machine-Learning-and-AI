```python
import pandas as pd

def extract_data(input_file,encoding):
    watched = []
    category = []

    with open(input_file, 'r', encoding=encoding) as file:
        lines = file.readlines()

    i = 0
    while i < len(lines):
        line = lines[i].strip()
        if line.startswith("Watched"):
            watched.append(line.split("Watched")[1].strip())
            category.append(lines[i+1].strip())
            i += 5  # Move to the next relevant section
        else:
            i += 1

    return watched, category


input_file = 'watch-history.txt'
watched, category = extract_data(input_file,'utf-8')

# Create a DataFrame
data = pd.DataFrame({'Watched': watched, 'Category': category})

# Print the DataFrame
print(data)

# If you want to save it to a CSV file
# data.to_csv('output.csv', index=False)

# if __name__ == "__main__":
#     main()

data.to_csv('output.csv', index=False)
```
